```python
import os
from PIL import Image
import cv2
import numpy as np

input_dir = 'F:/UoLby/by/images/train_smoke'
output_dir = 'F:/UoLby/by/images/train_smokeoutput'

os.makedirs(output_dir, exist_ok=True)

# Get all files in a directory
files = os.listdir(input_dir)

# Select all images
images = [file for file in files if file.lower().endswith(('png', 'jpg', 'jpeg', 'bmp'))]

def preprocess_image(image_path, output_path):
    # Use Pillow to read images
    image = Image.open(image_path)
    
    # Convert to grayscale 
    gray_image = image.convert('L')
    
    # Convert to numpy array
    image_np = np.array(gray_image)
    
    # Gaussian blur 
    blurred_image = cv2.GaussianBlur(image_np, (5, 5), 0)
    
    # Use Pillow to save preprocessed images
    output_image = Image.fromarray(blurred_image)
    output_image.save(output_path)

# Preprocess each image
for image_file in images:
    input_path = os.path.join(input_dir, image_file)
    output_path = os.path.join(output_dir, image_file)
    preprocess_image(input_path, output_path)
    print(f'Processed {input_path} and saved to {output_path}')

print('All images processed and saved.')
```

```python
import os
import shutil
import cv2
import numpy as np

input_dir = 'F:/UoLby/by/images/train_smokeoutput'
output_dir = 'F:/UoLby/by/images/image_segmentation'

# Clear the output directory
if os.path.exists(output_dir):
    shutil.rmtree(output_dir)
os.makedirs(output_dir, exist_ok=True)

# Defines the color range of the smoke in HSV color space
smoke_lower_bound = np.array([0, 0, 150])
smoke_upper_bound = np.array([180, 50, 255])

def segment_smoke(image_path, output_path):
    # Read the image and convert it to HSV color space                                 
    image = cv2.imread(image_path)
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    # Creating a smoke mask                                                           
    smoke_mask = cv2.inRange(hsv_image, smoke_lower_bound, smoke_upper_bound)
    
    # Use opening operation to remove small noise                                    
    kernel = np.ones((5,5), np.uint8)
    smoke_mask = cv2.morphologyEx(smoke_mask, cv2.MORPH_OPEN, kernel)
    
    # Extract smoke area                                                           
    segmented_image = cv2.bitwise_and(image, image, mask=smoke_mask)
    
    # Save the image                                                                 
    cv2.imwrite(output_path, segmented_image)

# Process images in the specified directory
image_files = [f for f in os.listdir(input_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]
for image_file in image_files:
    input_path = os.path.join(input_dir, image_file)
    output_path = os.path.join(output_dir, 'segmented_' + image_file)
    segment_smoke(input_path, output_path)

print("Image segmentation completed!")
```

```python
import cv2
import numpy as np
import os
import pandas as pd
from skimage.feature import greycomatrix, greycoprops

def extract_features(image):
    # Texture features using GLCM                                                     
    glcm = greycomatrix(image, [1], [0], 256, symmetric=True, normed=True)
    contrast = greycoprops(glcm, 'contrast')[0, 0]
    dissimilarity = greycoprops(glcm, 'dissimilarity')[0, 0]
    homogeneity = greycoprops(glcm, 'homogeneity')[0, 0]
    asm = greycoprops(glcm, 'ASM')[0, 0]
    energy = greycoprops(glcm, 'energy')[0, 0]
    correlation = greycoprops(glcm, 'correlation')[0, 0]
    
    # Shape characteristics                                                             
    contours, _ = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    areas = [cv2.contourArea(c) for c in contours]
    perimeters = [cv2.arcLength(c, True) for c in contours]
    
    # Color characteristics (grayscale intensity)                                        
    mean_intensity = np.mean(image)
    std_intensity = np.std(image)
    
    features = {
        'Contrast': contrast,
        'Dissimilarity': dissimilarity,
        'Homogeneity': homogeneity,
        'ASM': asm,
        'Energy': energy,
        'Correlation': correlation,
        'Max Area': max(areas) if areas else 0,
        'Max Perimeter': max(perimeters) if perimeters else 0,
        'Mean Intensity': mean_intensity,
        'Std Intensity': std_intensity
    }
    return features

def process_images(input_dir, output_dir):
    # Make sure the output directory exists
    os.makedirs(output_dir, exist_ok=True)
    features_list = []
    
    # Iterate over all image files in a directory
    for file in os.listdir(input_dir):
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(input_dir, file)
            image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            if image is not None:
                features = extract_features(image)
                features['Image'] = file  
                features_list.append(features)
    
    # Save feature data to a CSV file
    if features_list:
        df = pd.DataFrame(features_list)
        df.to_csv(os.path.join(output_dir, 'image_features.csv'), index=False)
        print("Features extracted and saved successfully.")

# Use functions to process images and save features
input_directory = 'F:/UoLby/by/images/image_segmentation'  
output_directory = 'F:/UoLby/by/images/feature_extraction'
process_images(input_directory, output_directory)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_curve, f1_score
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt

file_path = 'F:/UoLby/by/images/feature_extraction/output_features.csv' 
data = pd.read_csv(file_path)

energy_bins = np.linspace(data['Energy'].min(), data['Energy'].max(), 7)
area_bins = np.linspace(data['Max Area'].min(), data['Max Area'].max(), 7)

# Create new columns to store fire size and fire concentration level
data['Fire Size'] = pd.cut(data['Energy'], bins=energy_bins, labels=False, include_lowest=True)
data['Fire Concentration'] = pd.cut(data['Max Area'], bins=area_bins, labels=False, include_lowest=True)

# Ensure that the fire size and fire concentration columns are ranked from 1 instead of 0
data['Fire Size'] = data['Fire Size'] + 1
data['Fire Concentration'] = data['Fire Concentration'] + 1

# Preparing features and labels
X = data.drop(['Image', 'Fire Size', 'Fire Concentration'], axis=1)
y_size = data['Fire Size']
y_concentration = data['Fire Concentration']

# Training and prediction using random forest classifier                                                                         
model_size = RandomForestClassifier(n_estimators=100, random_state=42)
model_size.fit(X, y_size)

model_concentration = RandomForestClassifier(n_estimators=100, random_state=42)
model_concentration.fit(X, y_concentration)

# Binarize the labels                                                                                                                       
y_size_bin = label_binarize(y_size, classes=[1, 2, 3, 4, 5, 6])
y_concentration_bin = label_binarize(y_concentration, classes=[1, 2, 3, 4, 5, 6])

# Calculate the F1 score curve for each category                                                                                  
f1_scores_size = []
f1_scores_concentration = []

for i in range(6):
    y_size_pred = model_size.predict_proba(X)[:, i]
    precisions_size, recalls_size, thresholds_size = precision_recall_curve(y_size_bin[:, i], y_size_pred)
    f1_scores = 2 * (precisions_size * recalls_size) / (precisions_size + recalls_size)
    f1_scores_size.append((thresholds_size, f1_scores[:-1]))                                                                   

    y_concentration_pred = model_concentration.predict_proba(X)[:, i]
    precisions_concentration, recalls_concentration, thresholds_concentration = 
    precision_recall_curve(y_concentration_bin[:, i], y_concentration_pred)
    f1_scores = 2 * (precisions_concentration * recalls_concentration) / 
                (precisions_concentration + recalls_concentration)
    f1_scores_concentration.append((thresholds_concentration, f1_scores[:-1]))            
# Plot the total F1 score curve
plt.figure(figsize=(12, 6))

# Plot the total F1 score curve of Fire Size
plt.subplot(1, 2, 1)
for i, (thresholds, f1_scores) in enumerate(f1_scores_size, 1):
    plt.plot(thresholds, f1_scores, label=f"F1 Score for Fire Size Level {i}")
plt.xlabel("Threshold")
plt.ylabel("F1 Score")
plt.title("F1 Score Curve for Fire Size")
plt.legend()
plt.grid(True)

# Plot the total F1 score curve for Fire Concentration
plt.subplot(1, 2, 2)
for i, (thresholds, f1_scores) in enumerate(f1_scores_concentration, 1):
    plt.plot(thresholds, f1_scores, label=f"F1 Score for Fire Concentration Level {i}")
plt.xlabel("Threshold")
plt.ylabel("F1 Score")
plt.title("F1 Score Curve for Fire Concentration")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
